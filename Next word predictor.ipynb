Converting words to numbers

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
# input should be sent to a list, as there are multiple words
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
# this way unique tokens will get an index
```

```python
# now we will find out all the sentences present in our dataset
#for sentence in faqs.split('\n'):
#splitting the dataset on the basis of line change
  #print(sentence)
  #print(tokenizer.texts_to_sequences([sentence])[0])
```

now we need to generate inputs ad outputs, as in for the first sentence[92,1,13]-> [92] will be the input to which [1] should be the output, further [92,1] should be the input and [13] should be the output.

```python
input_sequences = []

for sentence in faqs.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1, len(tokenized_sentence)):
       input_sequences.append(tokenized_sentence[:i+1])
```

1. List to store input sequences
2. Generate input sequences for each sentence
3. for any sentence the loop will run for (no. of words - 1) times.
4. at first the loop runs for i=1, first index....then it will take i+1=2, run for first 2 index and so on.

```python
input_sequences
```

Now we will be zero padding the above output, this will add zeroes to every index, equal to the number of maximum words in any row.

```python
#finding length of every sentence and adding them to a list, then choosing the maximum one
max_length = max([len(x) for x in input_sequences])
```

```python
max_length
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_length, padding='pre')
```

```python
padded_input_sequences
```

now we can separate our input and output sequences, in each row -> input will be from (0 to (n-2)) column, and output will be (n-1)th column

```python
x = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

 as it is a classification problem, we will transform our output and replacing outputs with their one hot encoded representation -> if there are 9 words the representation for 2 will be [0 1 0 0 0 0 0 0 0]

```python
x.shape
```

```python
y.shape
```

```python
len(tokenizer.word_index)
```

```python
# we need to one hot encode y, as it is scalar for now
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes =284)      # num_classes = total number of words
# 284 because, one hot encoding starts from zero always, but here it is starting from 1 so we need to add 1 to the length
```

```python
y.shape
# every output is 284 number long sequence
```

```python
y
```

now we will build the architecture of our lstm model, 3layers(embedding, lstm, dense)

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()

model.add(Embedding(284, 100, input_length = 56))
model.add(LSTM(150))
model.add(Dense(284, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
model.summary()
```

In case of embedding - every input of 56 words, is given 100 numbers as output that is why weights is 284*100, this is conversion to 100D vector.
In case of LSTM - each 100D vector of a word will be input to the first timestep at t=1, then at t=2 second word will be the input.. this will go on till t=56, and the output will be of dimension(1,150)as the number of nodes in lstm is 150 so the dimension will be 150.
In case of dense - contains 284 nodes, which will give 284 output in the range of (0-1)(due to softmax which acts as probability), and the one with max probability will be our output.

```python
model.fit(x,y, epochs = 10)
# you can increase the number of epochs to 100 to increase the accuracy
```

```python
import time
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Assuming 'model' and 'tokenizer' are already defined and loaded
text = "total duration of"
num_predictions = 10
max_sequence_length = 56  # Adjust based on your model's input requirement

for i in range(num_predictions):
    # Tokenize
    token_text = tokenizer.texts_to_sequences([text])[0]
    print(f"Tokenized text: {token_text}")

    # Padding
    padded_token_text = pad_sequences([token_text], maxlen=max_sequence_length, padding='pre')
    print(f"Padded token text: {padded_token_text}")

    # Predict
    prediction = model.predict(padded_token_text)
    print(f"Prediction: {prediction}")
    pos = np.argmax(prediction, axis=-1)[0]  # Get the index of the predicted word
    print(f"Predicted position: {pos}")

    # Find the word corresponding to the predicted index
    predicted_word = None
    for word, index in tokenizer.word_index.items():
        if index == pos:
            predicted_word = word
            break

    if predicted_word:
        text = text + " " + predicted_word
        print(f"Updated text: {text}")
    else:
        print("Predicted word not found in the tokenizer's word index.")
        break

    time.sleep(1)
```
